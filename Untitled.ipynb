```python

dados = pd.read_excel('Imobiliario.xlsx')
```

```python
dados['Preco'].describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Qual a distribuição de preços de casas? (histograma)
dados['Preco'].hist(figsize=(5,3))
plt.xlabel("Preço")
plt.ylabel("Frequencia")
plt.show()
#Distribuição de preço de casas: Concentração de Preços: A maior parte dos imóveis parece estar concentrada em um intervalo de preços entre
# [2.000.000 - 6.000.000] Isso sugere que a maioria dos imóveis neste dados se enquadra nessa faixa de valor.
# Poucos Imóveis Caros: Observa-se uma queda acentuada na frequência de imóveis com preços superiores a 6.000.000.
```

```python
#Há relação entre preços de casas e a área da casa?
   #R- O gráfico de dispersão apresentado mostra uma relação positiva entre o preço e a área das casas.
   #Isso significa que, quanto maior a área de uma casa, maior tende a ser o seu preço.
plt.figure(figsize = (5,3))
plt.scatter(
    dados['Preco'],
    dados['Area'],
    c='red')

plt.xlabel("Preco")
plt.ylabel("Area")
plt.show()
```

```python
#A quantidade de quartos exerce influência no preço? E por número de banheiros?

#Tendência Crescente: À medida que o número de quartos aumenta, a mediana do preço também tende a aumentar.
#Isso indica que, imóveis com mais quartos são mais caros.
sns.catplot(data=dados, x='Quartos', y='Preco', kind='box')
plt.xlabel('Quartos')
plt.ylabel('Preço (R$)')
plt.title('Influência dos quartos no Preço')
plt.show()
```

```python
sns.catplot(data=dados, x='AquecimentoAgua', y='Preco', kind='box')
plt.xlabel('AquecimentoAgua')
plt.ylabel('Preço (R$)')
plt.title('Influência doAquecimentoAgua no Preço')
plt.show()

#bservamos que a mediana dos imóveis com aquecimento de água é superior àqueles sem aquecimento.
#Isso indica que, em média dos imóveis com aquecimento de água são mais caros.
# Esses outliers podem ser explicados por outros fatores como localização, tamanho do imóvel ou características especiais.
```

```python
sns.catplot(data=dados, x='Mobilia', y='Preco', kind='box')
plt.xlabel('Imóvel Mobiliado')
plt.ylabel('Preço (R$)')
plt.title('Influência do Mobiliado no Preço')
plt.show()
#Os imóveis não mobiliados apresentam, em média, os menores preços.
```

```python
sns.catplot(data=dados, x='NumAndares', y='Preco', kind='box')
plt.xlabel('NumAndares')
plt.ylabel('Preço (R$)')
plt.title('Influência do NumAndares no Preço')
plt.show()
```

```python
import seaborn as sns
#Area	Quartos	Banheiros	NumAndares VagasGaragem

matriz_correlacao = dados[['Preco','Area','Quartos','Banheiros','NumAndares', 'VagasGaragem']].corr()
# Criando o mapa de calor (heatmap) com Seaborn
sns.heatmap(matriz_correlacao, annot=True, cmap='coolwarm', linewidths=0.5)
```

```python

# Agrupar os dados por número de quartos e calcular a média do preço
df_grouped = dados.groupby('Banheiros')['Preco'].mean().reset_index()

sns.barplot(x='Banheiros', y='Preco', data=df_grouped)

```

```python
#Tendência crescente: À medida que o número de banheiros aumenta, a mediana do preço também aumenta.
#Isso indica que imóveis com mais banheiros são mais caros.
#Dispersão se tornam maiores à medida que o número de banheiros aumenta. Isso sugere que a variação de preços é maior em imóveis com mais banheiros.

sns.catplot(data=dados, x='Banheiros', y='Preco', kind='box')
plt.xlabel('Banheiros')
plt.ylabel('Preço (R$)')
plt.title('Influência dos Banheiros no Preço')
plt.show()
```
